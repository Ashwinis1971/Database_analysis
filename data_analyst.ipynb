```python
from sqlalchemy import create_engine, Column, Integer, String, ForeignKey, func, text, select
from sqlalchemy.orm import declarative_base, sessionmaker, relationship

# define the database engine

engine = create_engine('sqlite:///Jspider.db')
Base = declarative_base()

# define the tables as models

class Student(Base):
    __tablename__ = 'students'
    student_id = Column(Integer, primary_key=True)
    name = Column(String)
    class_name = Column(String, ForeignKey('classes.class_name'))
    score = Column(Integer)


class Class(Base):
    __tablename__ = 'classes'
    class_id = Column(Integer, primary_key=True)
    class_name = Column(String, unique=True)
    teacher = Column(String)

# Create the tables
Base.metadata.create_all(engine)

# create a session
Session = sessionmaker(bind=engine)
session = Session()


```

```python
query = session.query(Student.name, Student.class_name, Student.score)
results = query.all()
results
```
